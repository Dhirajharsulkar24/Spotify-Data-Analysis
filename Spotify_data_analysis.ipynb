# Data Analysis

# **Top 5 most popular artists**

```python
top_5_artists=df.groupby('artist').count().sort_values(by='song_title',ascending=False)['song_title'][:5]
print(top_5_artists)
```

```python
top_5_artists.plot.bar(title='Most Popular Artists based on Song Title',figsize=(8,8))
```

# **Top 5 loudest Tracks**

```python
top_5_loudest=df[['loudness','song_title']].sort_values(by='loudness',ascending=True)[:5]
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="loudness",y="song_title",data=top_5_loudest)
plt.title("Top 5 Loudest Tracks")
plt.show()
```

# **Artist with most danceability song**

```python
artist_most_danceability=df[['artist','danceability','song_title']].sort_values(by='danceability',ascending=False)[:5]
artist_most_danceability

```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="artist",y="danceability",data=artist_most_danceability)
plt.title("Artist VS Danceability")
```

# **Top 10 Instrumental Songs**

```python
top_10_instrumental_songs=df[['song_title','artist','instrumentalness']].sort_values(by='instrumentalness',ascending=False)[:5]
top_10_instrumental_songs
```

```python
plt.figure(figsize=(10,8))
plt.pie(x="instrumentalness",data=top_10_instrumental_songs,labels=top_10_instrumental_songs.song_title,autopct="%.2f%%")
plt.title("    Top 5  Instrumental Songs")
plt.show()
```

# **Multiple Feature Plot**

```python
df.columns
```

```python
intrest_feature=['tempo','loudness','acousticness','danceability',
                      'duration_ms','energy','instrumentalness','liveness','speechiness','valence']
```

```python
for i in intrest_feature:
  pos_data=df[df['target']==1][i]
  neg_data=df[df['target']==0][i]
  plt.figure(figsize=(12,7))

  sns.distplot(pos_data,bins=30,label="Positive",color="green")
  sns.distplot(neg_data,bins=30,label="Negative",color="blue")

  plt.legend(loc="upper right")
  plt.title(f"Positive and negative histogram for {i}")
  plt.show()
```
